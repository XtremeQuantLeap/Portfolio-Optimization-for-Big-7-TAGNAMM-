```python
log_rets = np.log(df/df.shift(1))
```

```python
log_rets.dropna()
```

```python
np.exp(log_rets.cumsum()).dropna()
```

```python
log_rets.cumsum().apply(np.exp).dropna()
```

```python
log_rets.cumsum().apply(np.exp).plot(figsize = (10,8), grid = True);
```
